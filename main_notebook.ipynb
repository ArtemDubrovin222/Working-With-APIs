```python
import requests

def get_http_cat(status_code):
    url = f"https://http.cat/{status_code}"
    response = requests.get(url)
    if response.status_code == 200:
        return response.content
    else:
        return None

status_code = 404
cat_image = get_http_cat(status_code)
if cat_image:
    with open(f"{status_code}.jpg", "wb") as f:
        f.write(cat_image)
    print(f"Cat image for status code {status_code} saved successfully.")
else:
    print(f"Failed to retrieve cat image for status code {status_code}.")
```

```python
#2
url2 = "https://www.cbr.ru/scripts/XML_daily.asp?date_req=02/03/2022"
response2 = requests.get(url2)
#print(response2.content)
```

```python
import xml.etree.ElementTree as ET
data_all = response2.text
#print(type(data_all))
```

```python
root = ET.fromstring(data_all)
# Create an empty list to store dictionaries
data = []
# Iterate through Valute elements and extract values
for entry in root.findall('Valute'):
    id = entry.attrib['ID']
    numCode = entry.find('NumCode').text
    charcode = entry.find('CharCode').text
    nominal = entry.find('Nominal').text
    name = entry.find('Name').text
    value = entry.find('Value').text
    vunitRate = entry.find('VunitRate').text
    
    # Append extracted data to list of dictionaries
    data.append({"ID": id, "NumCode": numCode, "CharCode": charcode, 
                 "Nominal": nominal, "Name": name, "Value": value, 
                 "VunitRate": vunitRate})
# Convert list of dictionaries to DataFrame
df = pd.DataFrame(data)
df
```
